```python
busqueda= 'san martin de valdeiglesias madrid'
busqueda= busqueda.replace(' ','-')


x= 1
ids= []

while True:
    url= f'https://www.idealista.com/venta-viviendas/{busqueda}/pagina-{x}.htm'
    browser.get(url)
    time.sleep(random.randint(10,12))
    
    try:
        browser.find_element("xpath",'//*[@id="didomi-notice-agree-button"]').click()
    except:
        pass

    
    html= browser.page_source
    
    soup= bs(html, 'lxml')
    
    pagina_actual= int(soup.find('main',{'id': 'main-content'}).find('div',{'class': 'pagination'}).find('li', {'class': 'selected'}).text)
    
    
    if x == pagina_actual:
        
        articles= soup.find('main', {'id': 'main-content'}).find_all('article')
    
    else:
        break
        
    x= x+1
    
    for article in articles:
        
        id_inmuebles= article.get('data-adid')
        
        ids.append(id_inmuebles)
        
        time.sleep(random.randint(1,3))
        
        print(id_inmuebles)
    
    ids= [muebles for muebles in ids if muebles is not None]
```

```python
id_inmuebles= pd.DataFrame(ids)
id_inmuebles.columns= ['id']
id_inmuebles
```

```python
id_inmuebles.to_csv('id_inmuebles_idealista-2024-03-30.csv', index=False)
```

```python
inmuebles= pd.Series()
```

```python
def parsear_inmuebles (id_inmuebles):
    
    print('\n Inmueble numero: ' + id_inmuebles)
    
    url= 'https://www.idealista.com/inmueble/'+ id_inmuebles + '/'
    
    browser.get(url)
    
    html= browser.page_source
    
    soup= bs(html, 'lxml')
    
    
    
    titulo= titulo= soup.find('span', {'class': 'main-info__title-main'}).text.strip()
    
    print('\n Titulo: '+ titulo)
    
    
    
    precio = int(soup.find('span', {'class': 'txt-bold'}).text.replace(',', '').replace('.', ''))
    
#     print('\n Precio: '+ precio)
    
    
    
    ubicacion= soup.find('span', {'class': 'main-info__title-minor'}).text.split(',')[0]
    
    print('\n Ubicacion: '+ ubicacion)
    
    
    
    direccion= soup.find_all('li', {'class': 'header-map-list'})
    direccion_completa= [a.text.strip() for a in direccion if a.text.strip() != '']
    
#     print('\n Direccion Completa: '+ direccion_completa)
    
    
    
    carac_ba= soup.find('div', {'class': 'details-property-feature-one'})
    caract_basicas= [a.text.strip() for a in carac_ba.find_all('li')]
    
#     print('\n Caracteristicas Basicas' + caract_basicas)
    
    
    
    equip_edif=soup.find('div', {'class': 'details-property-feature-two'}).find('div', {'class': 'details-property_features'})
    equipamiento_edificio= [a.text.strip() for a in equip_edif.find_all('li')]
    
#     print('\n Equipamiento/Edificio' + equipamiento_edificio)
    
    
    
    inmuebles['titulo']= titulo
    
    inmuebles['precio']= precio
    
    inmuebles['ubicacion']= ubicacion
    
    inmuebles['direccion']= direccion_completa
    
    inmuebles['caracteristicas_basicas']= caract_basicas
    
    inmuebles['caracteristicas_extras']= equipamiento_edificio
    
    
    df_inmuebles= pd.DataFrame(inmuebles)
    
    return (df_inmuebles.T)
```

```python
df_inmuebles= parsear_inmuebles(id_inmuebles.iloc[0].id)

for i in range(1, len(ids)):
    
    df_inmuebles= pd.concat([df_inmuebles, parsear_inmuebles(ids[i])])
    
    time.sleep(random.randint(4, 8))
```
